## <span style='color:Green'>Activity 8: </span> `Using a Scatter Plot to Visualize Correlation between Various Animals`

```python
import pandas as pd

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

plt.rcParams['figure.figsize'] = (14, 6);
```

```python
an = pd.read_csv('./Activity 8/Datasets/anage_data.csv')
```

```python
an
```

```python
longevity = 'Maximum longevity (yrs)'
mass = 'Body mass (g)'
an = an[np.isfinite(an[longevity]) & np.isfinite(an[mass])]

amphibia = an[an['Class'] == 'Amphibia']
aves = an[an['Class'] == 'Aves']
mammalia = an[an['Class'] == 'Mammalia']
reptilia = an[an['Class'] == 'Reptilia']
```

```python
plt.figure(figsize=(16, 6), dpi=300)

plt.scatter(amphibia[mass], amphibia[longevity], label='Amphibia')
plt.scatter(aves[mass], aves[longevity], label='Aves')
plt.scatter(mammalia[mass], mammalia[longevity], label='Mammalia')
plt.scatter(reptilia[mass], reptilia[longevity], label='Reptilia')

plt.legend()

ax = plt.gca()
ax.set_xscale('log')
ax.set_yscale('log')

plt.xlabel('Body mass in grams')
plt.ylabel('Maximum longevity in years')

plt.show()
```

## <span style='color:Green'>Activity 9: </span> `Working on Radar Charts`

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
"""
Sample data

Attributes: Efficiency, Quality, Commitment, Responsible Conduct, Cooperation


"""
data = pd.DataFrame({'Employee': ['Alex', 'Alice', 'Chris','Jennifer'], 'Efficiency': [5, 4, 4, 3],
'Quality': [5, 5, 3, 3], 'Commitment': [5, 4, 4, 4],
'Responsible Conduct': [4, 4, 4, 3],
'Cooperation': [4, 3, 4, 5]})
```

```python
data
```

```python
attributes = list(data.columns[1:])
values = list(data.values[:, 1:])
employees = list(data.values[:, 0])
angles = [n / float(len(attributes)) * 2 * np.pi for n in range(len(attributes))]

angles += angles[:1]
values = np.asarray(values)
values = np.concatenate([values, values[:, 0:1]], axis=1)
```

```python
plt.figure(figsize=(8, 8), dpi=250)

for i in range(4):
    ax = plt.subplot(2, 2, i + 1, polar=True)
    ax.plot(angles, values[i])
    ax.set_yticks([1, 2, 3, 4, 5])
    ax.set_xticks(angles)
#     ax.set_xticklabels(attributes)
    ax.set_title(employees[i], fontsize=14, color='r')

plt.tight_layout()

plt.show()
```

## <span style='color:Green'>Activity 11: </span> `Plotting Multiple Images in a Grid`

```python
import matplotlib.image as mpimg
import os
```

```python
img_filenames = os.listdir('./Activity 8/Datasets/images')
imgs = [mpimg.imread(os.path.join('./Activity 8/Datasets/images', img_filename)) for img_filename in img_filenames]
```

```python
plt.figure(figsize=(16, 8), dpi=250)

fig, axes = plt.subplots(2, 2, figsize=(15, 12))
labels = ['coast', 'beach', 'building', 'city at night']
for i in range(2):
    for j in range(1):
        axes[i,j].imshow(imgs[i])
        axes[i,j].set_xticks([])
        axes[i,j].set_yticks([])
        axes[i,j].set_xlabel(labels[i])
        
for i in range(2):
    for j in range(1):
        axes[i,j+1].imshow(imgs[i+2])
        axes[i,j+1].set_xticks([])
        axes[i,j+1].set_yticks([])
        axes[i,j+1].set_xlabel(labels[i+2])
```
